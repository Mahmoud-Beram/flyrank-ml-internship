## 2. The question: decision, action, cost of a wrong call

**Decision:** which content pages should be reviewed and refreshed first,
given that an editor only has time for a limited number per week.

**Action:** an editor opens the top-ranked pages from the list and
updates or rewrites them.

**Cost of a wrong call:** if a page is flagged that isn't really
declining, the editor wastes limited time on it instead of a page that
actually needs help. If a genuinely declining page is missed, it keeps
losing traffic unnoticed until the damage is bigger.

**Why data/ML helps:** with 30,000 pages and only a handful matching a
simple rule (e.g. "stale AND still visible" caught only 17 pages), a
plain rule is too narrow to rank the rest usefully. The signals that
matter (age, freshness, position, CTR, impressions) interact in ways a
single hand-written condition can't capture well.
